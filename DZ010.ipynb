Реализация через get_dummies

```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
pd.get_dummies(data['whoAmI'])
```

1. Реализация через OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
data_new = onehotencoder.fit_transform(data.values)
pd.DataFrame(data_new.toarray(),
columns=onehotencoder.categories_).head(10)
```

2. Реализация через category_encoders

```python
df_catenc = data.copy()
 
import category_encoders as ce
 
# в параметр cols передадим столбцы, которые нужно преобразовать
ohe_encoder = ce.OneHotEncoder(cols = ['whoAmI'])
# в метод .fit_transform() мы передадим весь датафрейм целиком
df_catenc = ohe_encoder.fit_transform(df_catenc)
list(df_catenc)
df_catenc.rename(columns = {'whoAmI_1':'human', 'whoAmI_2':'robot'}, inplace = True )
list(df_catenc)
df_catenc.head()
```
